```python
import sys
sys.path.append("../ARCH_package")

import basic

import dill
import numpy as np

import plotly.graph_objects as go

# Import non-synonymous mutations as exported in LiFT.py
with open('../Exports/LBC_non-synonymous_fitted_bayesian.dill', 'rb') as infile:
    lbc = dill.load(infile)
''
cohort = [traj for part in lbc for traj in part.trajectories if traj.fitness >0.02]

from tqdm import tqdm
vaf_range = np.linspace(0.001, 0.02, 1000)
time_range = np.linspace(0, 10, 10_000)

def fitness_evolution_low_std(t, fitness = 0.1, vaf = 0.02):

    mean_term = vaf*np.exp(fitness*t)
    var_term = vaf*(2*1.3+fitness)*np.exp(fitness*t)*(np.exp(fitness*t)-1)/(fitness*2*100_000)

    return mean_term - 2*np.sqrt(var_term)

def sequencing_high_std(vaf = 0.02, resolution = 2_000):

    mean_term = vaf
    var_term = vaf*(1-vaf)/resolution

    return mean_term + 2*var_term

fig = go.Figure()
fitness_labels = ['5%', '10%', '15%', '20%']
fitness_range = [0.05, 0.1, 0.15, 0.2]

for fitness, fitness_label in zip(fitness_range, fitness_labels):
    time_to_next_observation = []
    for vaf in vaf_range:
        fluctuations = fitness_evolution_low_std(time_range,
                                                 fitness=fitness,
                                                 vaf=vaf)
        sequencing = sequencing_high_std(vaf=vaf)
        difference = fluctuations - sequencing

        positive_elements = np.where(difference > 0)
        if positive_elements[0].size!= 0:
            fluctuations_time = time_range[positive_elements[0][0]]
            time_to_next_observation.append(fluctuations_time)
    
    fig.add_trace(
            go.Scatter(x=time_to_next_observation,
                        y=vaf_range,
                        name= fitness_label))

fig.update_layout(title='Time to next clinical observation',
                  xaxis_range=[0,10],
                  xaxis_title='Time',
                  yaxis_title='Vaf')

fig
```

```python
time_range = np.linspace(0, 20, 10_000)
vaf_range = np.linspace(0.01, 0.5, 100)

def fluctuations_difference(t, fitness= 0.1, vaf = 0.1):

    mean_term = vaf*(np.exp(fitness*t)-1)
    var_term = vaf*(2+fitness)*np.exp(fitness*t)*(np.exp(fitness*t)-1)/(fitness*2*100_000)

    return mean_term - 2*np.sqrt(var_term)

time_to_next_observation = []
for vaf in vaf_range:
    fluctuations = fluctuations_difference(time_range, vaf=vaf)
    absolute_val_array = np.abs(fluctuations - 0)
    smallest_difference_index = absolute_val_array.argmin()

    time_to_next_observation.append(time_range[smallest_difference_index])

fig = go.Figure()
fig.add_trace(
    go.Scatter(x=time_to_next_observation,
               y=vaf_range)
)
fig.update_layout(title=('Time to next clinical observation<br>'
                          'Fixed mutation fitness = 0.1'),
                  xaxis_title='Time',
                  yaxis_title='VAF')
```
